***Question 8 et 9°)***

```python
def forward_selection(X_train, Y_train, seuil_p_value, nb_p_value_to_plot) :
    
    features = list(range(0, round(var_explicatives,3)))
    
    #initialisation de toutes les variables 
    features_sel = {}
    p_values_sel = {}
    t_stat_max = 0
    t_stat_list = []
    stop_indexer = []
    i = 0
    
    
    while len(features) > 0:
       
        taille = len(Y_train)
        
        #initialisation de epsilon et t_stat
        epsilon = []
        t_stat = []
             
        #Pour chaque feature dans la liste features
        for feature in features:
            
            #On effectue une regression linéaire OLS
   
            MODEL_OLS.fit(X_train[[feature]], Y_train)
            
            #Prediction du modèle
            Y_hat = MODEL_OLS.predict(X_train[[feature]])
            
            #Variance des résidus 
            sig_res = ((Y_train - Y_hat)**2).sum()/(taille - 2)
            
            # Variance de la feature
            sig_coef = np.sqrt(sig_res / ((X_train[feature] - X_train[feature].mean())**2).sum())
            
            #Calcul de la stat de test
            stat_test = float(abs(MODEL_OLS.coef_[0][0]/sig_coef))
            
            #Rajouter le residu a la liste des residus
            epsilon.append(Y_train - Y_hat)
            
            #Ajout de la valeur de la stat pour le calcul du maximum de la t_stat de la feature
            t_stat.append(stat_test)
            t_stat_list.append(round(stat_test,3))    
                
        #Récupérer l'indice correspondant au max de la stat de test, et la valeur de l'index correspondant
        t_stat_max = max(t_stat)
        index_max = t_stat.index(t_stat_max)
        p_value_max = (1 - stat.t.cdf(abs(t_stat_max), taille - 2)) * 2
       
        #Si p-value > seuil, on stop le process
        if p_value_max > seuil_p_value:
            stop_indexer.append(i)
            
        i +=1
            
        #Ajout de la p_value max
        p_values_sel[features[index_max]] = p_value_max
        
        #Ajout la valeur max de la t_stat
        features_sel[features[index_max]] = round(t_stat_max, 3)
        
        
        Y_train =  epsilon[index_max]
        #On supprime la colonne qui a été ajoutée à features_sel afin de ne pas boucler a l'infinie   
        del features[index_max]
    
    #lists = sorted(p_values_sel.items()) #Ordonné par clé, retourne une liste de tuples
    x, y = zip(*sorted(p_values_sel.items())) #Extrait les valeurs de la liste de tuples

    
    print("Voici les variables selectionnées lors du forward selection ainsi que les t-test correspondantes : \n \n ")
    
    #print(features_sel)
    print({k: features_sel[k] for k in list(features_sel)[:min(stop_indexer)]})


    plt.figure(figsize=(8,8))
    plt.plot(sorted(y[:nb_p_value_to_plot]))
    
    plt.axhline(y=0.1, color='b', linestyle='-')
    
    return features_sel, t_stat_list
```

```python
feature_sel, tstat_list = forward_selection(X_train, Y_train, 0.1, 50);
```

```python
# Création d'une fonction permettant de déterminer le alpha optimal de la régression de RIDGE.
def ridge(data, alpha):
    clf = lm.Ridge()
    #initialisation du vecteur de risque de prédiction
    risque_predict = []
    
    #Bouclage sur les valeurs de alpha définis précédements.
    for a in alpha :
        
        clf.set_params(alpha=a)
        risk_predict_CV = []
        
        for train_index, test_index in kf.split(data):
            
            risque_ridge = 0
            risque_predict_mean = 0
            
            clf.fit(data.iloc[train_index], Y_train.iloc[train_index])
            
            Y_hat = clf.predict(data.iloc[test_index])
            Y_test_CV = Y_train.iloc[test_index]
            
            risque_ridge = mean_squared_error(Y_test_CV, Y_hat)
            risk_predict_CV.append(risque_ridge)
            
            
        risque_predict.append(np.mean(risk_predict_CV))
        
    index_optimal = np.argmin(risque_predict)  
    alpha_optimal = alpha[index_optimal]

    plt.figure(figsize=(8,6))
    plt.plot(alpha, risque_predict)
    plt.axvline(x=alpha_optimal, linewidth=2, color = 'black', label ='Alpha optimal ≈' + str(round(alpha_optimal,3)))
    plt.xlabel('alpha')
    plt.ylabel('Risque de validation croisé')
    plt.tight_layout()
    plt.legend(loc=7)
    plt.show() 
    print("Le alpha optimal du modèle RIDGE est %.3f" %alpha_optimal)
```

```python
#Création de la liste des alphas pour lesquels nous allons effectuer la regression de ridge. 
alpha = np.logspace(np.log10(0.00001), np.log10(1000))
```
